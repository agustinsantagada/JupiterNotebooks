### Filtramos el dataset

```python
df_mach_learn =df[(df["property_type"] == "PH") | (df["property_type"] == "Casa") | (df["property_type"] == "Departamento") ]
df_mach_learn = df_mach_learn[(df_mach_learn.l2 == "Capital Federal")]
df_mach_learn = df_mach_learn[(df_mach_learn.surface_total >= 15)]
df_mach_learn = df_mach_learn[(df_mach_learn.surface_total <= 1000)]
df_mach_learn = df_mach_learn[(df_mach_learn.price <= 4000000)]
df_mach_learn = df_mach_learn[["rooms","bedrooms","bathrooms","surface_total","surface_covered","price"]]
df_mach_learn.isna().sum()
df_mach_learn = df_mach_learn.dropna()
df_mach_learn.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import mean_squared_error
```

### Definimos las variables

```python
X=df_mach_learn[['surface_total']]
y=df_mach_learn['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3 , random_state=42)
```

### Entrenamos el √°rbol

```python
tree=DecisionTreeRegressor(max_depth=3,random_state=1)
tree.fit(X_train,y_train)
```

```python
y_train_pred=tree.predict(X_train)
```

```python
y_train_pred
```

```python
y_test_pred=tree.predict(X_test)
```

```python
y_test_pred
```

```python
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
print(rmse_train) 
print(rmse_test)
```

### Vamos a analizar el error del √°rbol de dicisi√≥n y vecinos cercanos

```python
k_nei=[1,2,3,4,5,6,7,8,9] 
k_nei_train_error=[]
k_nei_test_error=[]

for k in k_nei:   
    nei=KNeighborsRegressor(n_neighbors=k)
    nei.fit(X_train,y_train)

    y_train_pred_k=nei.predict(X_train)
    y_test_pred_k=nei.predict(X_test)

    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred_k))
    rmse_test= np.sqrt(mean_squared_error(y_test, y_test_pred_k))
    print(rmse_train) 
    print(rmse_test)
    
    k_nei_train_error.append(rmse_train)
    k_nei_test_error.append(rmse_test)
    
    print("\nCon un depth de:", k)
    print(f'Ra√≠z del error cuadr√°tico medio en Train: {rmse_train}')
    print(f'Ra√≠z del error cuadr√°tico medio en Test: {rmse_test}')

```

```python
k_tree=[1,2,3,4,5,6,7,8,9] 
k_tree_train_error=[]
k_tree_test_error=[]

for k in k_tree:   
    tree=DecisionTreeRegressor(max_depth=k,random_state=1)
    tree.fit(X_train,y_train)

    y_train_pred=tree.predict(X_train)
    y_test_pred=tree.predict(X_test)

    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test= np.sqrt(mean_squared_error(y_test, y_test_pred))
    print(rmse_train) 
    print(rmse_test)
    
    k_tree_train_error.append(rmse_train)
    k_tree_test_error.append(rmse_test)
    
    print("\nCon un depth de:", k)
    print(f'Ra√≠z del error cuadr√°tico medio en Train: {rmse_train}')
    print(f'Ra√≠z del error cuadr√°tico medio en Test: {rmse_test}')
```

## RMSE

Analizamos los valores con   RMSE ya que es la mejor medida frente  MAE debido a que nos da la diferencia entre la predicci√≥n y el valor real del error 

```python

fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True,figsize=(15,10))

ax1.plot(k_tree,k_tree_train_error,'o-',label='train')
ax1.plot(k_tree,k_tree_test_error,'o-',label='test')
ax2.plot(k_nei,k_nei_train_error,'o-',label='train')
ax2.plot(k_nei,k_nei_test_error,'o-',label='test')

plt.legend()
ax1.set_title("Modelo de arbol de decision", size =14)
ax2.set_title("Modelo de vecinos cercanos", size =14)
ax1.set_xlabel("N¬∞ ramas")
ax2.set_xlabel("N¬∞ vecinos")
ax1.set_ylabel("Error(En USD)")
ax2.set_ylabel("Error(En USD)")
plt.show()
```

#### Vemos que el resultado mas conveniente es el de arbol de decisi√≥n pero la diferencia de acierto es de 200 mil d√≥lares, debido al grango de precios de 4 millones.
